```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Model comparison Table
comparison_df = pd.DataFrame({
    "Model": [
        "Isolation Forest",
        "LightGBM",
        "Autoencoder",
        "Fusion Model"
    ],

    "Accuracy": [
        0.49,
        0.9999,
        0.99,
        0.97
    ],

    "Purpose": [
        "Baseline Anomaly Detection",
        "Known Attack Classification",
        "Zero-Day Detection",
        "Hybrid Detection System"
    ]
})

comparison_df
```

```python
# Accuracy Bar Chart
plt.figure(figsize=(8,5))

plt.bar(
    comparison_df["Model"],
    comparison_df["Accuracy"]
)

plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.ylim(0, 1.05)

plt.xticks(rotation=10)

plt.show()
```

```python
# Save comaprision table
BASE_PATH = "/content/drive/MyDrive/Zero-Day-IoT-Project"

comparison_df.to_csv(
    BASE_PATH + "/04_Reports/model_comparison.csv",
    index=False
)

print("Comparison Table Saved")
```

```python
# Precision / Recall / F1 Comparison
metric_df = pd.DataFrame({
    "Metric": ["Precision", "Recall", "F1-Score"],

    "LightGBM": [1.00, 1.00, 1.00],

    "Autoencoder": [0.99, 1.00, 1.00],

    "Fusion": [0.96, 0.98, 0.97]
})

metric_df
```

```python
# Zero-Day Detection Comparison Graph

import matplotlib.pyplot as plt

models = ['Isolation Forest', 'Autoencoder', 'Fusion Model']

zero_day_accuracy = [0.49, 0.99, 0.97]

plt.figure(figsize=(8,5))

bars = plt.bar(models, zero_day_accuracy)

plt.ylim(0, 1.05)

plt.ylabel("Accuracy")
plt.xlabel("Models")

plt.title("Zero-Day Attack Detection Accuracy Comparison")

# Add accuracy labels
for bar, acc in zip(bars, zero_day_accuracy):
    plt.text(
        bar.get_x() + bar.get_width()/2,
        acc + 0.01,
        f"{acc*100:.1f}%",
        ha='center'
    )

plt.show()
```
